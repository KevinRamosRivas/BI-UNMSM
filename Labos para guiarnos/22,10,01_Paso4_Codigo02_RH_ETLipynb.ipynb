```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
query = 'SELECT distinct(Education) FROM data10 order by Education '
df = pysqldf(query)
df.head(18)
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
query = 'SELECT Education FROM data10 '
df = pysqldf(query)
df.head(18)
```

```python
# Casting del dato 'Education', conversión del dato
for i in range(len(df)):
    if df.iloc[i,0]==1:
       df.iloc[i,0]=0
    elif df.iloc[i,0]==2:
       df.iloc[i,0]=1
    elif df.iloc[i,0]==3:
       df.iloc[i,0]=2
    elif df.iloc[i,0]==4:
       df.iloc[i,0]=3
    elif df.iloc[i,0]==5:
       df.iloc[i,0]=4
```

```python
data10.drop(["Education"], axis = 1, inplace = True)
data = pd.concat([data10,df],axis=1)
data.head(200)

```
